### 10. Write a short report on the visual observations of the data.

````python
import matplotlib.pyplot as plt

df[['age', 'income']].hist(bins=20, figsize=(8,4))
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

corr_vars = df[['income', 'age', 'visits', 'hospital', 'chronic']]

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_vars.corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Matrix of Healthcare Variables')
plt.show()
````

````python
df['IncomeGroup'] = pd.qcut(
    df['income'],
    4,
    labels=['Low', 'Lower-Mid', 'Upper-Mid', 'High']
)
````

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

df.boxplot(column='visits', by='IncomeGroup', ax=axes[0])
df.boxplot(column='hospital', by='IncomeGroup', ax=axes[1])
df.boxplot(column='chronic', by='IncomeGroup', ax=axes[2])

axes[0].set_title('Physician Visits')
axes[1].set_title('Hospital Stays')
axes[2].set_title('Chronic Conditions')

for ax in axes:
    ax.set_xlabel('Income Group')
    ax.set_ylabel('Count')

plt.suptitle('Healthcare Utilization by Income Group')
plt.show()
````

````python
plt.figure(figsize=(8, 5))

plt.scatter(df['age'], df['visits'], alpha=0.4)

plt.xlabel('Age')
plt.ylabel('Physician Visits')
plt.title('Physician Visits by Age')

plt.show()
````

````python
sns.regplot(x='age', y='visits', data=df, scatter_kws={'alpha':0.3})
````

````python
health_visits = df.groupby('health')['visits'].mean()

plt.figure(figsize=(8, 5))
health_visits.plot(kind='bar')

plt.xlabel('Self-Reported Health Status')
plt.ylabel('Average Physician Visits')
plt.title('Average Visits by Health Status')

plt.show()
````
